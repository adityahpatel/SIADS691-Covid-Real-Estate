```python
import numpy as np
# import modin.pandas as pd # This is for when we have all the files
import pandas as pd  # if modin is not needed
import seaborn as sns
import pickle
import plotly.express as px
import os
```

```python
# Read in New York Dictionary of dataframes
ny_dict_df = pd.read_pickle("datasets/NY_data/final_data_frame/NY_dict_df_col_cleaned.pkl")
print('done')
```

# Let's look at NYC because it is the largest county by population

```python
ny_dict_df['New York'] = ny_dict_df['New York'].sort_values(by = 'transaction_date', ascending=True)
```

```python
for county in ny_dict_df.keys():
    # convert sale date column to datetime
    ny_dict_df[county]['transaction_date'] = pd.to_datetime(ny_dict_df[county]['transaction_date'])
```

```python
# show average values
ny_dict_df['New York'].mean()
```

```python
ny_dict_df['New York'].groupby(by='transaction_date').count()
```

```python
# let's graph the trendline for sales count
daily_sales_scatter = px.scatter(ny_dict_df['New York'].groupby(by='transaction_date').count(),y='type',  trendline="lowess").update_layout(yaxis_title="Sales Count", xaxis_title="Transaction Date")
daily_sales_scatter.show()
```
